```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df_input)
data_scaled
```

```python
features = data_scaled
target = data_scaled[:, 0]
```

```python
TimeseriesGenerator(features, target, length=6, sampling_rate=1, batch_size=1)[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=120, shuffle= False)
print('X_train.shape: ', x_train.shape)
print('X_test.shape: ', x_test.shape)
```

```python
win_length = 144 # 1 day = 144 of data, 5 day = 720 of data
batch_size = 32
num_features = 6 
train_generator =  tf.keras.preprocessing.sequence.TimeseriesGenerator(x_train, y_train, length=win_length, sampling_rate=1, batch_size=batch_size)
test_generator =  tf.keras.preprocessing.sequence.TimeseriesGenerator(x_test, y_test, length=win_length, sampling_rate=1, batch_size=batch_size)
```

```python
train_generator[0]
```

```python
model4 = tf.keras.Sequential()
model4.add(tf.keras.layers.InputLayer((win_length, num_features)))
model4.add(tf.keras.layers.Conv1D(64, kernel_size=2))
model4.add(tf.keras.layers.LSTM(128, return_sequences=True))
model4.add(tf.keras.layers.LeakyReLU(alpha=0.5))
model4.add(tf.keras.layers.Dropout(0.3))
model4.add(tf.keras.layers.LSTM(64, return_sequences=True))
model4.add(tf.keras.layers.Flatten())
model4.add(tf.keras.layers.Dense(8, 'relu'))
model4.add(tf.keras.layers.Dense(1, 'sigmoid'))

model4.summary()

# model4 = tf.keras.Sequential()
# model4.add(tf.keras.layers.LSTM(128, input_shape=(win_length, num_features), return_sequences=True))
# model4.add(tf.keras.layers.LeakyReLU(alpha=0.5))
# model4.add(tf.keras.layers.LSTM(128, return_sequences=True))
# model4.add(tf.keras.layers.LeakyReLU(alpha=0.5))
# model4.add(tf.keras.layers.Dropout(0.3))
# model4.add(tf.keras.layers.LSTM(64, return_sequences=False))
# model4.add(tf.keras.layers.Dropout(0.3))
# model4.add(tf.keras.layers.Dense(1, activation='sigmoid'))
# model4.summary()
```

```python
cp2 = ModelCheckpoint('../save/model4/', save_best_only=True)
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=2,
    mode='min'
)
model4.compile(loss=tf.keras.losses.MeanSquaredError(),
              optimizer= tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=[tf.keras.metrics.MeanAbsoluteError()] )
```

```python

history = model4.fit_generator(
    train_generator, epochs=20,
    validation_data=test_generator,
    shuffle=False,
    callbacks=[early_stopping, cp2
    ])
```

```python
model4.evaluate_generator(
    test_generator, verbose=0
)
```

```python
predictions = model4.predict_generator(test_generator)
```
